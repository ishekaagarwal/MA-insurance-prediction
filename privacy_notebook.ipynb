# Project Background and Motivation

[Return to contents](#contents)

This project’s aim is to contribute to the new initiative of the State of Massachusetts to help build machine learning models to predict which residents are not protected by health care insurance. Especially during the COVID-19 pandemic, without proper health insurance, the residents hesitate to seek medical care if they contract the virus as they will likely receive large medical bills. Therefore, it is crucial for us to build a prediction model that would identify MA residents that are not currently protected, so that the State government can reach out to these individuals and inform them of the no-cost plans provided by the State. We fit, compare, and contrast a range of models, including a Multivariate Logistic Regression (baseline), Decision Tree, Random Forest, and FFNN.

Because the raw U.S. Census data includes sensitive personally identifiable information, including names and addresses, it is essential to preserve the privacy of people in the training set after the model's deployment. Therefore, we make use of a differentially private machine learning library in this project, SmartNoise, in training our models.

### Variables of interest, notes, managing missing entries, and corresponding transformations:
We selected a number of predictor variables which we will convert to the format suitable for our analysis. 

- `AGEP`: age (0-99), renamed to `Age`
- `SEX`: 1 for Male, 2 for Female. Convert to `Sex_Female`, subtract 1, obtaining a binary indicator for whether a subject is Female. Mean of `Sex_Female` will correspond to proportion of females in the set.
- `CIT`: citizenship status. Convert to `Not_US_Citizen` where `CIT == 5`
- `NATIVITY`: convert to `Foreign_born` where `NATIVITY == 2` (Foreign born)
- `ENG`: ability to speak English. Convert to `Poor_English` indicator where `ENG == 3` ('Not well') or `ENG == 4` ('Not at all')
- `RAC1P`: race code. Convert to `White` where `RAC1P == 1`; `Black` where `RAC1P == 2`; `Asian` where `RAC1P == 6`; `Native` where `RAC1P == 3` or `RAC1P == 4` or `RAC1P == 5` or `RAC1P == 7`;
- `HISP`: Hispanic origin. Convert to `Hispanic` where `HISP != 1`
- `Racial_Minority` where `RAC1P != 1` or `HISP != 1`
- `MAR`: marital status. Convert to `Married` indicator where `MAR == 1`;
- `DIS`: disability. Convert to `Disability` indicator where `DIS == 1` (With a disability)
- `SCHL`: educational attainment. Convert to `Secondary_Ed` indicator where `SCHL >=16`, meaning person received high school diploma; `Post_Secondary_Ed` where `SCHL >=18`, meaning person attended some college (even if for < 1 year); `Tertiary_Ed` where `SCHL >=20`, meaning person obtained at least Associate's degree; `Professional_Ed` where `SCHL >=22`, meaning person obtained a Master's degree or higher
- `WAGP`: wages or salary income past 12 months. `NA` for individuals < 15 y.o.; rename to `Wage`
- `INTP`: interest, dividends, and net rental income past 12 months; bottom-coded to loss of `$10,000` and top-coded to income of `$999,999`; rename to `Int_Divs_Rent_Income`
- `PAP`: public assistance income in past 12 months. Rename to `Public_Assistance_Income`. Add indicator `High_Public_Assistance` where `Public_Assistance_Income > 5000`
- `PERNP`: total person's earnings: bottom-coded at loss of `$10,000` and top-coded at `$1,999,998`; rename to `Total_Personal_Earnings`.
- `ESR`: employment status. Convert to `Unemployed` indicator variale where `ESR==3`; `Not_Labor_Force` indicator where `ESR==6`
- `POVPIP`: income-to-poverty ratio. Convert to also `25_Poverty_Threshold`, `5_Poverty_Threshold`, `1_Poverty_Threshold` indicators that an individual is in the bottom 25, 5, 1% in terms of income to povery threshold ratio
- `JWMNP`: travel time to work, top-coded to `200` minutes. Convert to `Long_commute` where `JWMNP > 45`

### Response:
- `HICOV`: rename to `Not_Insured` indicator where `HICOV == 2`

### Managing missing entries:
Additionally, many observations (individuals) in our dataset have certain variables missing. We cannot use rows with missing data to fit our models. However, dropping all rows with at least one variable missing would not be appropritate, because that could potentially introduce a bias into our data. Therefore, we need to consider  how to treat each missing variable individually, and make careful decisions about dropping or imputation at the same time as performing data transformations specified above.

However, we found that most missing entries occur where subjects are < 16 y.o. children. We decided to drop children < 16 y.o. from our dataset, because for them a lot of variables are missing (wage, employment hours, etc.). Additionally, it is likely that children's insurance depends wholly on their parents' insurance coverage. And the state is not going to be reaching out to 12 year olds, so dropping children will help us streamline our analysis.

## Data Balancing

Our dataset is extremely imbalanced, with only 2.4% of individuals not being protected by insurance. This can potentially negatively impact our models' performance because the models won't be able to extract meaningful relationships between variables. We will explore 3 ways of balancing our data to fix this problem: 
- upsampling the minority class (uninsured people), 
- downsampling the majority class (insured people),
- generating synthetic samples. 

We will fit our target models (logistic regression, decision tree, and FFNN) to the original data and then compare how much the performance improves for each resampling technique, to determine which approach delivers the most improvement. We will pay most attention to improvement in recall and F1 scores, because, as we established earlier, these are our preferred metrics.

It is important to note, however, that we split our data into train and test in the beginning, and only attempt to resample and balance the training set that's used in training our models. Otherwise, the same observations can end up in both train and test sets (in case of oversampling). By splitting train/test beforehand, we will make sure that throughout the entire process, we always test our models on the same subset of the original (imbalanced) dataset, thus making the results generalizable to the real world.

Once we determine which data balancing approach performs the best, we will extend our analysis to creating differentially private versions of our target models.

## Decision Tree/Random Forest

We built the decision tree models on three types of datasets (upsampled, downsampled, synthetic). Please note that **we did not build a differentially private model for decision trees** because we only used decision trees to identity the best tree depth for the Random Forest.

And, **to establish the best max depth for the decision tree model, we built models for various tree depths, and compared the mean accuracy over 5-fold cross validation**. We are performing 5-fold cross validation and not a higher number to ensure that we don’t end up with folds that have very few minority class examples because they are already too few to begin with.Additionally, we ensured that the cross validation folds are stratified i.e. they have the same proportion of majority and minority classes. This was done to ensure that at least each fold is comparable.

As we can be seen below, we identified that the decision tree model with:
- **upsampled** dataset has the best performance when the maximum depth is **30**.
- **downsampled** dataset has the best performance when the maximum depth is **30**. 
- **synthetic** dataset has the best performance when the maximum depth is **22**.

Please note that after seeing the performance of the logistic regression model on the original dataset, we decided not to continue with the original dataset. This is because the higher F1-Scores we received for the other upsampled, downsampled or synthetically-upsampled dataset validated our hypothesis that these balanced datasets are much better for model-training.

```python
depths = list(range(1, 40))

def calc_mean(x_train, y_train, depths):
    cvmeans = []
    train_scores = []
    cv_scores = []
    
    for depth in depths:
        clf = DecisionTreeClassifier(max_depth=depth)

        train_scores.append(clf.fit(x_train, y_train).score(x_train, y_train))
        scores = cross_val_score(clf, x_train, y_train, cv=5, scoring='accuracy')

        cvmeans.append(scores.mean())

    return cvmeans, train_scores
```

```python
cvmeans_upsampled, train_scores_upsampled = calc_mean(upsampled_x_train, upsampled_y_train, depths)
cvmeans_upsampled = np.array(cvmeans_upsampled)

cvmeans_downsampled, train_scores_downsampled = calc_mean(downsampled_x_train, downsampled_y_train, depths)
cvmeans_downsampled = np.array(cvmeans_downsampled)

cvmeans_synth, train_scores_synth = calc_mean(synth_x_train, synth_y_train, depths)
cvmeans_synth = np.array(cvmeans_synth)
```

### Random Forest with Privacy
For the differentially-private Random Forest, we used Sam Fletcher and Md Zahidul Islam's work published in the paper: 
Fletcher, S., & Islam, M. Z. (2017). Differentially private random decision forests using smooth sensitivity. Expert Systems with Applications, 78, 16-31.

In the paper, the authors present an algorithm that builds a differentially-private random forest model. The authors of the algorithm tried multiple epsilon values (0.01, 0.1, 1) for identifying the epsilon value that results in the best performance. As can be seen in the graph below, through their analysis, they identified that epsilon value of 1 produces the best prediction accuracy, but of course it also reduces the privacy of the model. On the other end, an epsilon value of 0.01 gives poorer performance, but ends up preserving a lot more privacy. So, an **epsilon value of 0.1 seems to be a happy-medium**. As seen in the graph, prediction accuracy at epsilon value of 0.1 is not too far off from the accuracy at epsilon value of 1, especially at dataset size of 30K and 300K which is the range that the size of our dataset falls in.

And, for the **number of trees we chose is 64** because the paper shows that there is "sweet spot at 30 to 100 trees where prediction accuracy is highest". We specifically chose number of trees as 64 because we didn't want to have too few or too many trees, and additionally going a bit lower helped us save a lot of computational time.

We also read another paper: Lee, J., & Clifton, C. (2011, October). How much is enough? choosing ε for differential privacy. In International Conference on Information Security (pp. 325-340). Springer, Berlin, Heidelberg. This survey paper on differential privacy further confirms that a low epsilon value can usually be good at privacy. More specifically, the paper say that "given sufficiently low ε, ε-differential privacy does limit an adversary’s ability to identify an individual:.

Please note that all of the code for the differentially-private Random Forest model was from Fletcher and Islam's paper. The GitHub repository for their code: https://github.com/sam-fletcher/Smooth_Random_Trees

## FFNN
To implement the feed forward neural network we used the pytorch library. We implemented a model with 4 fully connected dense layers: the first one with 320 output nodes, the second one with 160, the third one with 80 and the last one with 2 (corresponding to the 2 classes of insured and uninsured people). For the internal layers we used tanh activation while for the output layer we used softmax activation to return the two classes.


```python
device = torch.device('cuda' if torch.cuda.is_available else 'cpu')
print(device)
```

```python
def get_DataLoader(x_df, y_df, batch_size=32):
    features = torch.tensor(np.array(x_df), dtype=torch.float)
    targets = torch.tensor(np.array(y_df), dtype=torch.long)

    ds = torch.utils.data.TensorDataset(features, targets)
    ds_loader = torch.utils.data.DataLoader(ds, batch_size=batch_size, shuffle=True, drop_last=True)

    return ds_loader
```

```python
training_loader = get_DataLoader(synth_x_train, synth_y_train, batch_size=32)
validation_loader = get_DataLoader(x_test, y_test, batch_size=32)
```

```python
class FFNN(nn.Module):
    def __init__(self):
        super().__init__()

        self.fc1 = nn.Linear(28,320)
        self.fc2 = nn.Linear(320,160)
        self.fc3 = nn.Linear(160,80)
        self.fc4 = nn.Linear(80,2)
           
    def forward(self, x):
        x = F.tanh(self.fc1(x)) 
        x = F.tanh(self.fc2(x))
        x = F.tanh(self.fc3(x))
        x = F.softmax(self.fc4(x))
        return x
```

### FFNN with Privacy
To add privacy to our feed forward neural network implemented above, we used the opacus library. This library implements a privacy engine that takes a series of parameters to make the optimizer differentially private. When building a differentially private feed forward neural network we had to take into consideration the following parameters :

- Noise_multiplier: this noise parameter determines the amount of noise that is added to the gradients before they are applied to the optimizer. The noise_multiplier parameter can go from 0 to infinity. Higher values mean a higher level of privacy but are more challenging for the training process and therefore impact accuracy. Lower values of the noise_multiplier correspond to lower levels of privacy but higher accuracy as a result of the data being less noisy and more close to the original data. We chose a noise_multiplier of 0.65 because it provides a level of privacy without losing too much accuracy. In fact, our validation accuracy went down to 0.684 in the differentially private FFNN from the 0.869 we got in the FFNN without privacy.
- L2_norm_clip: this parameter clips the gradient and limits the sensitivity of the optimizer to individual training points. Therefore, the risk that the model memorizes individual data points is reduced. For the gradient clipping parameter we used 2.0.
- Batch_size: this parameter indicates the amount of samples used per update of the weights during training. This applies to all neural network based trainings and is not specific to differential privacy. In our model we use a batch_size of 32.
- Num_microbatches: the size of the microbatches should divide the value we used for batch_size. We chose to keep the microbatch size equal to batch_size
